```python
import matplotlib.pyplot as plt
import numpy as np
import math
from sympy import *
```

График функции `f(x) = r - aru^3` 

```python
r = 1
for a in range(1,11,1):
        x = np.arange(0,2,0.01)
        y = r - a*r*x**3
        z = np.sqrt(x)
        idx = np.argwhere(np.diff(np.sign(y - z))).flatten()
        if a == 10:
            plt.plot(x,y,label='g(u) = r -aru^3')
            plt.plot(x,z,label='f(u) = sqrt(u)')
            plt.plot(x[idx], y[idx],'ro',label='solution')
            plt.legend()
        else:
            plt.plot(x,y)
            plt.plot(x,z)
            plt.plot(x[idx], y[idx],'ro',)
        plt.axis([0,2,0,2])
        plt.xlabel('u')
        plt.ylabel('f(u)')
        plt.title('Graphic f(u), r = 1, a in [1,10]')
#plt.legend('f(u) = r -aru^3','f(u) = sqrt(u)')
```
